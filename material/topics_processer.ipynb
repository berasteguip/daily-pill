```python
import pandas as pd
import glob
```

```python
file_names = [name for name in glob.glob("topics/*.csv")]
file_names_clean = [name.split("\\")[1].split(".")[0] for name in file_names]
print(file_names_clean)
```

```python
for name in file_names_clean:
    
    df = pd.read_csv(f"topics/{name}.csv")
    df['Estado'] = 'pending'
    df.columns = ['id', 'Titulo', 'Contenido', 'Estado']
    df.to_csv(f"topics/{name}.csv", index=False)
```
